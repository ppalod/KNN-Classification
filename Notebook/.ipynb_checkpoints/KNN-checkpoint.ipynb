```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import collections
```

#### Functions used in the homework

```python
def plot_error_rate(result):
    plt.figure(figsize=(10,5))
    plt.grid(True)
    plt.plot(result["k_values"],result["test_errors"],color='red',label='Test error')
    plt.xlabel('K values')
    plt.ylabel('Test Error rate')
    plt.title('Error for different K values')

def best_estimates(result):
    best_train_error = np.amin(result["train_errors"])
    best_test_error = np.amin(result["test_errors"])
    index_of_best_error = np.where(np.array(result["test_errors"]) == best_test_error)[0][0]
    best_k_value = result["k_values"][index_of_best_error]
    print("Best train error:",best_train_error)
    print("Best test error:",best_test_error)
    print("Best k value (k*):",best_k_value)

def calculate_weighted_distance(p_value):
    knn = KNeighborsClassifier()
    result = collections.OrderedDict(
        [
            ("k_values",[]),
            ("train_scores",[]),
            ("train_errors",[]),
            ("test_scores",[]),
            ("test_errors",[])
        ])
    for k in range(1,197,5):
        knn.n_neighbors=k
        knn.p=p_value
        knn.weights='distance'
        knn.fit(X_train, Y_train)
        train_score=knn.score(X_train,Y_train)
        train_error=1-train_score
        y_pred = knn.predict(X_test)
        test_score=accuracy_score(Y_test,y_pred)
        test_error=1-test_score
        result["k_values"].append(k)
        result["train_scores"].append(train_score)
        result["train_errors"].append(train_error)
        result["test_scores"].append(test_score)
        result["test_errors"].append(test_error)
    plot_error_rate(result)
    best_test_error = np.amin(result["test_errors"])
    index_of_best_error = np.where(np.array(result["test_errors"]) == best_test_error)[0][0]
    best_k_value = result["k_values"][index_of_best_error]
    return (best_test_error,best_k_value)
```

### (a) Download the Vertebral Column Data Set

####  Load data

```python
df = pd.read_csv('../Data/vertebral_column_data/column_2C.dat',sep="\s+",header=None)
df.columns = ['pelvic_incidence','pelvic_tilt','lumbar_lordosis_angle','sacral_slope','pelvic_radius','grade_of_spondylolisthesis','class']
df
```

### (b) Pre-Processing and Exploratory data analysis:

#### Change the datapoint values in column "class" (Replace "NO" with 0 and "AB" with 1

```python
df.loc[df['class']=='NO','class']=0
df.loc[df['class']=='AB','class']=1
df
```

#### Class 0 and Class 1 df

```python
Class_0 = df.loc[df['class']==0]
Class_1 = df.loc[df['class']==1]
```

#### i. Make scatterplots of the independent variables in the dataset. Use color to show Classes 0 and 1.

```python
pair_plot = sns.pairplot(df,hue='class');
pair_plot.fig.suptitle("Relation among pelvic_incidence, pelvic_tilt, lumbar_lordosis_angle, sacral_slope, pelvic_radius and grade_of_spondylolisthesis", y=1.05);
plt.show()
```

#### ii. Make boxplots for each of the independent variables. Use color to show Classes 0 and 1

```python
for idx in range(0,6):
    sns.boxplot(x=df['class'],y=df[df.columns[idx]],data=df)
    plt.show()
```

#### iii. Select the first 70 rows of Class 0 and the first 140 rows of Class 1 as the training set and the rest of the data as the test set.

#### Create training set

```python
train_df1 = Class_0[:70] 
train_df2 = Class_1[:140] 
train_df = pd.concat((train_df1,train_df2))
train_df
```

#### Create test set

```python
test_df1 = Class_0[70:] 
test_df2 = Class_1[140:] 
test_df = pd.concat((test_df1,test_df2))
test_df
```

#### Split training set into X_train, Y_train and test set into X_test, Y_test

```python
X_train = train_df.iloc[:, :-1]
Y_train = train_df.iloc[:,-1]
X_test = test_df.iloc[:, :-1]
Y_test = test_df.iloc[:,-1]

# to remove Unknown label type: 'unknown' error

X_train = X_train.astype('float')
Y_train = Y_train.astype('int')
X_test = X_test.astype('float')
Y_test = Y_test.astype('int')

# print(X_train)
# print(Y_train)
# print(X_test)
# print(Y_test)
```

### (c) Classification using KNN on Vertebral Column Data Set

#### i. Write code for k-nearest neighbors with Euclidean metric (or use a software package)

```python
knn = KNeighborsClassifier(n_neighbors=5) # For 5 nearest neighbors
knn.metric='euclidean'
knn.fit(X_train, Y_train)
y_pred = knn.predict(X_test)

# Train Accuracy and error

print("Training accuracy is:",knn.score(X_train,Y_train))
print("Training error is:",1-knn.score(X_train,Y_train))

# Test Accuracy and error

print("Test accuracy is:",accuracy_score(Y_test,y_pred))
print("Test error is:",1-accuracy_score(Y_test,y_pred),'\n')

# Confusion matrix for test data

print("Confusion matrix for test data:\n",confusion_matrix(Y_test, y_pred),'\n')

# Classification report for test data

print("Classification report for test data:\n",classification_report(Y_test, y_pred))
```

#### ii. Test all the data in the test database with k nearest neighbors. Take decisions by majority polling. Plot train and test errors in terms of k for k belongs to {208, 205,...,7,4,1} (in reverse order). You are welcome to use smaller increments of k. Which k* is the most suitable k among those values? Calculate the confusion matrix, true positive rate, true negative rate, precision, and F1-score when k = k*

#### Classify data and calculate best k

```python
knn = KNeighborsClassifier()
result = collections.OrderedDict(
    [
        ("k_values",[]),
        ("train_scores",[]),
        ("train_errors",[]),
        ("test_scores",[]),
        ("test_errors",[])
    ])
for k in range(208,0,-3):
    knn.n_neighbors=k
    knn.p=2 # Euclidean distance
    knn.fit(X_train, Y_train)
    train_score=knn.score(X_train,Y_train)
    train_error=1-train_score
    y_pred = knn.predict(X_test)
    test_score=accuracy_score(Y_test,y_pred)
    test_error=1-test_score
    result["k_values"].append(k)
    result["train_scores"].append(train_score)
    result["train_errors"].append(train_error)
    result["test_scores"].append(test_score)
    result["test_errors"].append(test_error)
    
best_estimates(result)
```

#### Plot train and test errors in terms of k

```python
plt.figure(figsize=(10,5))
plt.grid(True)
plt.plot(result["k_values"],result["train_errors"],color='blue',label='Train error')
plt.plot(result["k_values"],result["test_errors"],color='red',label='Test error')
plt.xlabel('K values')
plt.ylabel('Error rate')
plt.legend(loc='lower right')
plt.title('Error for different K values')
```

#### Calculate the confusion matrix, true positive rate, true negative rate, precision, and F1-score when k = k* (best_k)

```python
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, Y_train)
y_pred = knn.predict(X_test)

cf_mx = confusion_matrix(Y_test, y_pred)
print("Confusion matrix for test data: \n",cf_mx,'\n')

TN, FP, FN, TP = confusion_matrix(Y_test,y_pred).ravel()

TPR = TP/(TP+FN)
print("True positive rate: ",TPR,'\n')

TNR = TN/(TN+FP)
print("True negative rate: ",TNR,'\n')

precision = precision_score(Y_test,y_pred)
print("Precision: ",precision,'\n')

recall = recall_score(Y_test,y_pred)
print("Recall", recall,'\n')

F1_score = f1_score(Y_test,y_pred)
print("F1-score: ",F1_score,'\n')
```

#### iii. Since the computation time depends on the size of the training set, one may only use a subset of the training set. Plot the best test error rate, which is obtained by some value of k, against the size of training set, when the size of training set is N belongs to {10,20,.....,210}. Note: for each N, select your training set by choosing the first N/3 rows of Class 0 and the first N - N/3 rows of Class 1 in the training set you created in 1(b)iii. Also, for each N, select the optimal k from a set starting from k = 1, increasing by 5. For example, if N = 200, the optimal k is selected from {1,6,11,....,196}  This plot is called a Learning Curve.

```python
final_result = collections.OrderedDict(
    [
        ("N",[]),
        ("test_error_rate",[])
    ])

for N in range(10,220,10):
    class_0_train = train_df1[:int(N/3)]
    class_1_train = train_df2[:int(N-(N/3))]
    class_0_test = Class_0[int(N/3):]
    class_1_test = Class_1[int(N-(N/3)):]
    train_data = pd.concat((class_0_train,class_1_train))
    test_data = pd.concat((class_0_test,class_1_test))
    
    X_train = train_data.iloc[:, :-1]
    Y_train = train_data.iloc[:,-1]
    X_test = test_data.iloc[:, :-1]
    Y_test = test_data.iloc[:,-1]

    # to remove Unknown label type: 'unknown' error

    X_train = X_train.astype('float')
    Y_train = Y_train.astype('int')
    X_test = X_test.astype('float')
    Y_test = Y_test.astype('int')
    
    knn = KNeighborsClassifier()
    result = collections.OrderedDict(
        [
            ("k_values",[]),
            ("train_scores",[]),
            ("train_errors",[]),
            ("test_scores",[]),
            ("test_errors",[])
        ])
    
    for k in range(1,N+1,5):
        knn.n_neighbors=k
        knn.p=2
        knn.fit(X_train, Y_train)
        train_score=knn.score(X_train,Y_train)
        train_error=1-train_score
        y_pred = knn.predict(X_test)
        test_score=accuracy_score(Y_test,y_pred)
        test_error=1-test_score
        result["k_values"].append(k)
        result["train_scores"].append(train_score)
        result["train_errors"].append(train_error)
        result["test_scores"].append(test_score)
        result["test_errors"].append(test_error)
        
    best_test_error = np.amin(result["test_errors"])
    
    final_result["N"].append(N)
    final_result["test_error_rate"].append(best_test_error)
```

#### Best test error rate plot

```python
plt.figure(figsize=(10,5))
plt.grid(True)
plt.plot(final_result["N"],final_result["test_error_rate"],color='blue')
plt.xlabel('N values')
plt.ylabel('Best test error rate')
plt.title('Learning Curve')
```

#### i. Minkowski Distance:

#### A. which becomes Manhattan Distance with p = 1.

```python
knn = KNeighborsClassifier()
result = collections.OrderedDict(
    [
        ("k_values",[]),
        ("train_scores",[]),
        ("train_errors",[]),
        ("test_scores",[]),
        ("test_errors",[])
    ])
for k in range(1,197,5):
    knn.n_neighbors=k
    knn.p=1
    knn.fit(X_train, Y_train)
    train_score=knn.score(X_train,Y_train)
    train_error=1-train_score
    y_pred = knn.predict(X_test)
    test_score=accuracy_score(Y_test,y_pred)
    test_error=1-test_score
    result["k_values"].append(k)
    result["train_scores"].append(train_score)
    result["train_errors"].append(train_error)
    result["test_scores"].append(test_score)
    result["test_errors"].append(test_error)
    
best_estimates(result)
plot_error_rate(result)
```

#### B. with log10(p) belongs to {0.1, 0.2, .... , 1}. In this case, use the k* you found for the Manhattan distance in 1(d)iA. What is the best log10(p)?

```python
result = collections.OrderedDict(
    [
        ("p_values",[]),
        ("train_scores",[]),
        ("train_errors",[]),
        ("test_scores",[]),
        ("test_errors",[])
    ])
for p_i in np.arange(0.1,1.1,0.1):
    knn.n_neighbors=6
    knn.p=10**p_i
    knn.fit(X_train, Y_train)
    train_score=knn.score(X_train,Y_train)
    train_error=1-train_score
    y_pred = knn.predict(X_test)
    test_score=accuracy_score(Y_test,y_pred)
    test_error=1-test_score
    result["p_values"].append(p_i)
    result["train_scores"].append(train_score)
    result["train_errors"].append(train_error)
    result["test_scores"].append(test_score)
    result["test_errors"].append(test_error)

best_test_error = np.amin(result["test_errors"])
index_of_best_error = np.where(np.array(result["test_errors"]) == best_test_error)[0][0]
best_p_value = result["p_values"][index_of_best_error]

print("Best log10(p) value:",best_p_value)
print("Best p value:",10**best_p_value)
```

#### C. which becomes Chebyshev Distance with p -> inf

#### ii. Mahalanobis Distance

```python
knn = KNeighborsClassifier()
result = collections.OrderedDict(
    [
        ("k_values",[]),
        ("train_scores",[]),
        ("train_errors",[]),
        ("test_scores",[]),
        ("test_errors",[])
    ])
cov = np.cov(X_train, rowvar =False)
for k in range(1,197,5):
    knn.n_neighbors=k
    knn.metric='mahalanobis'
    knn.metric_params={'VI':cov}
    knn.fit(X_train, Y_train)
    train_score=knn.score(X_train,Y_train)
    train_error=1-train_score
    y_pred = knn.predict(X_test)
    test_score=accuracy_score(Y_test,y_pred)
    test_error=1-test_score
    result["k_values"].append(k)
    result["train_scores"].append(train_score)
    result["train_errors"].append(train_error)
    result["test_scores"].append(test_score)
    result["test_errors"].append(test_error)

best_estimates(result)
plot_error_rate(result)
```
